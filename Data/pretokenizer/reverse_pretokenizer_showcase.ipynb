```python
import ast

from pretokenizer import pretokenize
from reverse_pretokenizer import reverse
```

```python
examples = open("code_examples", "r").read().split("\n\n\n\n")
```

without dedent

```python
for example in examples:
    parsed = pretokenize(ast.parse(example), _use_dedent=False)
    print(f"example:\n{example}\n\nparsed:\n{reverse(parsed, _use_dedent=False)}\n\n")
```

with dedent

```python
for example in examples:
    parsed = pretokenize(ast.parse(example), _use_dedent=True)
    print(f"example:\n{example}\n\nparsed:\n{reverse(parsed, _use_dedent=True)}\n\n")
```
