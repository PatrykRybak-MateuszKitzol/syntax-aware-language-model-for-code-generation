```python
import ast
from pretokenizer import pretokenize
```

```python
examples = open("code_examples", "r").read().split("\n\n\n\n")
```

```python
for example in examples:
    parsed = pretokenize(ast.parse(example), _use_dedent=True, _use_semantics=True) 
    print(f"example:\n{example}\n\nparsed:\n{parsed}\n\n")
```
